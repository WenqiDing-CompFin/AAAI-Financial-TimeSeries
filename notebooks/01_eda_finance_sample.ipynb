# Exploratory analysis — sample finance panel

Quick look at the synthetic daily multi-asset dataset used for local demos.

```python
from pathlib import Path
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

root = Path("../data/sample/finance_daily")
if not (root / "time_series.csv").exists():
    import sys
    sys.path.append("..")
    from src.data.sample_data import generate_sample_finance_dataset
    generate_sample_finance_dataset(root)

df = pd.read_csv(root / "time_series.csv", parse_dates=["date"], index_col="date")
blob = np.load(root / "dataset.npz", allow_pickle=True)
labels = blob["labels"]
df.head()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(10, 6), sharex=False)
df[["sp500", "nikkei", "ftse"]].div(df.iloc[0]).plot(ax=axes[0], title="Normalized equity indices")
pd.Series(labels).value_counts().sort_index().plot(
    kind="bar", ax=axes[1], title="Label counts (0=down, 1=neutral, 2=up)"
)
plt.tight_layout()
plt.show()
```
